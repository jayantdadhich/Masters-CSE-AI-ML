Splitting the dataset 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

Model 1-Linear Regression

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score #Taking different types of Training and Testing
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
print(mse)
mean_mse=np.mean(mse)
print(mean_mse)
```

Fitting the model

```python
lin_reg.fit(X_train,y_train)
y_pred = lin_reg.predict(X_test)
y_pred
```

Performance metrics

```python
from sklearn.metrics import r2_score
r2_score1=r2_score(y_pred,y_test)
print(r2_score1)
```

```python
Accuracy=r2_score1*100
print(" Accuracy of the model is %.2f" %Accuracy)
```

Visualization

```python
plt.scatter( y_test,y_pred,color = 'red')
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()
```

Model 2-Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100,max_features = 'sqrt', max_depth = 5, random_state = 18)
# Training the model
regressor.fit(X_train, y_train)
#Prediction on Test Data
test_data_prediction = regressor.predict(X_test)
print(test_data_prediction)
```

Performance metrics

```python
from sklearn import metrics
# R squared Error
error_score = metrics.r2_score(y_test, test_data_prediction)
print("R squared error : ", error_score)
```

```python
Accuracy=error_score*100
print(" Accuracy of the model is %.2f" %Accuracy)
```

Visualization

```python
y_test = list(y_test)
plt.plot(y_test, color = "blue", label = "Actual Value")
plt.plot(test_data_prediction, color = "green", label = "Predicted Value")
plt.title("Actual Price vs Predicted Price")
plt.xlabel("No. of values")
plt.ylabel("House Price")
plt.legend()
plt.show()
```
